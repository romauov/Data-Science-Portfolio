##   FROM

1.  -40   ,     2 .      ( title),   ( rental_rate),   ( length)    ( rating).

```SQL

SELECT title, rental_rate, length, rating
FROM movie
WHERE rental_rate > 2
ORDER BY length DESC
LIMIT 40
;

```

2.       .      :
*   ( rating);
*      ( length);   min_length  max_length ;
*    ( length);   avg_length;
* ,       ( rental_rate);   min_rental_rate, max_rental_rate, avg_rental_rate .

     .

```SQL

SELECT rating,
    MIN(length) AS min_length,
    MAX(length) AS max_length,
    AVG(length) AS avg_length,
    MIN(rental_rate) AS min_rental_rate,
    MAX(rental_rate) AS max_rental_rate,
    AVG(rental_rate) AS avg_rental_rate
FROM (SELECT title,
           rental_rate,
           length,
           rating
    FROM movie
    WHERE rental_rate > 2
    ORDER BY length DESC
    LIMIT 40) AS best_rating
GROUP BY rating
ORDER BY AVG(length)
;

```

3.    ,         .     .   avg_min_length  avg_max_length .

```SQL

SELECT AVG(netop.min_length) AS avg_min_length,
        AVG(netop.max_length) AS avg_max_length
FROM
    (SELECT top.rating,
       MIN(top.length) AS min_length,
       MAX(top.length) AS max_length,
       AVG(top.length) AS avg_length,
       MIN(top.rental_rate) AS min_rental_rate,
       MAX(top.rental_rate) AS max_rental_rate,
       AVG(top.rental_rate) AS avg_rental_rate
FROM
  (SELECT title,
          rental_rate,
          length,
          rating
   FROM movie
   WHERE rental_rate > 2
   ORDER BY length DESC
   LIMIT 40) AS top
GROUP BY top.rating
ORDER BY avg_length) AS netop
;

```

    perc    .       month    .

```SQL

WITH 
y_12 AS (SELECT EXTRACT(MONTH FROM CAST(invoice_date AS timestamp)) AS month,
        SUM(total) AS sum_total_2012
        FROM invoice
        WHERE EXTRACT(YEAR FROM CAST(invoice_date AS timestamp)) = 2012
        GROUP BY month
        ORDER BY month
        ),
y_13 AS (SELECT EXTRACT(MONTH FROM CAST(invoice_date AS timestamp)) AS month,
        SUM(total) AS sum_total_2013
        FROM invoice
        WHERE EXTRACT(YEAR FROM CAST(invoice_date AS timestamp)) = 2013
        GROUP BY month
        ORDER BY month
        )       
SELECT y_12.month AS month,
    y_12.sum_total_2012 AS sum_total_2012,
    y_13.sum_total_2013 AS sum_total_2013,
     ROUND(100 * (y_13.sum_total_2013 - y_12.sum_total_2012) / y_12.sum_total_2012) AS perc
FROM y_12
JOIN y_13 ON y_12.month = y_13.month
;

```

```SQL

SELECT f.name AS name_of_fund,
        c.name AS name_of_company,
        fr.raised_amount AS amount
FROM fund AS f
JOIN investment AS i ON f.id = i.fund_id
JOIN funding_round AS fr ON i.funding_round_id = fr.id
JOIN company AS c ON fr.company_id = c.id
WHERE c.milestones > 6
        AND fr.funded_at BETWEEN '2012-01-01' AND '2013-12-31'
;

```

```SQL

SELECT buyer.name AS buyer_company,
        a.price_amount AS amt,
        cb.name AS bought_company,
        cb.funding_total AS invst,
        ROUND(a.price_amount / cb.funding_total)
FROM company AS buyer
JOIN acquisition AS a ON buyer.id = a.acquiring_company_id
JOIN company AS cb ON a.acquired_company_id = cb.id

WHERE a.price_amount > 0 AND cb.funding_total > 0
ORDER BY amt DESC, bought_company

LIMIT 10;

```

23.          ,    ,   2011, 2012  2013 .         .        2011     .

```SQL

WITH y_11 AS (
    SELECT country_code,
        AVG(funding_total)
FROM company 
WHERE EXTRACT(YEAR FROM founded_at) = 2011
GROUP BY country_code),

y_12 AS (
    SELECT country_code,
        AVG(funding_total)
FROM company 
WHERE EXTRACT(YEAR FROM founded_at) = 2012
GROUP BY country_code),

y_13 AS (
    SELECT country_code,
        AVG(funding_total)
FROM company 
WHERE EXTRACT(YEAR FROM founded_at) = 2013
GROUP BY country_code)

SELECT  y_11.country_code,
        y_11.avg AS avg_11,
        y_12.avg AS avg_12,
        y_13.avg AS avg_13
FROM y_11
JOIN y_12 ON y_11.country_code = y_12.country_code
JOIN y_13 ON y_11.country_code = y_13.country_code

ORDER BY avg_11 DESC
;

```
